```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display, HTML
```

```python
credits = pd.read_csv("tmdb_5000_credits.csv")
movies_df = pd.read_csv("tmdb_5000_movies.csv")
```

```python
credits_rename = credits.rename(columns={'movie_id': 'id'}, index={'ONE': 'Row_1'})
credits_rename.head()
```

```python
merged = pd.merge(left=credits_rename, right=movies_df, how='left', left_on='id', right_on='id')
merged.head(2)
```

```python
df = merged.drop(columns=['homepage','title_x','title_y','status','production_countries'])
df.head(2)
```
